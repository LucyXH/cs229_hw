```python
import numpy as np
import pandas as pd

data_x = pd.read_table('logistic_x.txt', delim_whitespace=True, names=('x1', 'x2'),
                   dtype={'x1': np.float64, 'x2': np.float64}).values

data_y = pd.read_table('logistic_y.txt', delim_whitespace=True, names=('y'),
                   dtype={'y': np.float64}).values.reshape((data_y.shape[0],))

data_x = np.hstack( (data_x, np.ones((data_x.shape[0], 1))) ) 
```

```python
def J(seita):
    return np.sum(np.log (np.exp(- data_x.dot(seita) * data_y) + 1) ) / data_x.shape[0]
```

```python
def g(seita):
    return 1 / (np.exp(- data_x.dot(seita) * data_y) + 1)
```

```python
def dJ(seita):
    return - np.sum( np.vstack((1 - g(seita)) * data_y) * data_x, axis=0) / data_x.shape[0]
```

```python
def H(seita):
    return data_x.transpose().dot(np.vstack( g(seita) * (1 - g(seita)) * data_y * data_y) * data_x) / data_x.shape[0]
```

```python
seita = np.array([0,0,0])
for i in range(10):
    d_seita = np.linalg.inv(H(seita)).dot(dJ(seita))
    seita = seita - d_seita
    print i, seita, J(seita)
```
